```python
#import library
from sklearn.datasets import load_iris

df = pd.read_csv("/kaggle/input/iris-decision-tree/Iris.csv")

df.head()
```

```python
#memisahkan atribut label
X = df[["SepalLengthCm","SepalWidthCm","PetalLengthCm","PetalWidthCm"]]
y = df[['Species']]
```

```python
#memulai desision tree
from sklearn.tree import DecisionTreeClassifier
```

```python
#membuat model decision tree

tree_model = DecisionTreeClassifier()

#mencoba model
tree_model.fit(X,y)
```

```python
#mencoba modelnya

tree_model.predict([[5.1,3.4,1.4,0.2]])
```

```python
from sklearn.model_selection import train_test_split  
X_train, X_test, y_train, y_test= train_test_split(X, y, test_size= 0.25, random_state=100)
```

```python
from sklearn.model_selection import GridSearchCV

params = {'criterion': ['entropy', 'gini'], 'max_depth': [5, 6, 7, 8, 9, 10], 'min_samples_leaf': [1, 3, 5, 7, 9]}
gs = GridSearchCV(DecisionTreeClassifier(), params, cv=10, n_jobs=-1, verbose=1)
gs.fit(X_train, y_train)
```

```python
best_classifier = gs.best_estimator_
best_classifier
```

```python
from sklearn import tree

fig = plt.figure(figsize=(10,7))
tree.plot_tree(best_classifier, feature_names=df.columns, filled=True)
plt.show()
```
